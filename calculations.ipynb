```python
from math import *
from ase.structure import graphene_nanoribbon
import ASE_utils

max_no_reps = 5
ganrs = [graphene_nanoribbon(x, y, type='armchair', saturated=True) for x in range(1,max_no_reps+1) for y in range(1,x)]
gznrs = [graphene_nanoribbon(x, y, type='zigzag', saturated=True) for x in range(1,max_no_reps+1) for y in range(1,x)]

#ase default means that if we view the image we don't see the plane we want (i.e. we see a line of atoms)
for ribbon in ganrs+gznrs:
    ribbon.rotate('x',pi/2)
```

Deciding to use initial armchair structure 

```python
from IPython.core.display import Image
from ase.visualize import view

init_5_5 = graphene_nanoribbon(6, 5, type='armchair', saturated=True)
init_5_5.rotate('x',pi/2)
init_5_5.rotate('z',pi/2)

init_5_5.write('init_5_5.png')
Image(filename='init_5_5.png')
#view(init_5_5)
```

We will now export this to an xyz file and then use avogadro to cut it into fragments

```python
init_5_5.write('graphene_template.xyz','xyz')
```

```python
from ASE_utils import cut, multicut
import numpy as np
d=np.linalg.norm(init_5_5[62].position - init_5_5[44].position)

cut1 = [20, 62, +0.5]
cut2 = [15, 53, +0.5]
cut3 = [60, 42, +0.5]
cut4 = [111, 89,+0.5]
new_slice = multicut(init_5_5, *zip(cut1,cut2,cut3,cut4))

new_slice.write('new_slice.png')
Image(filename='new_slice.png')
```

```python
cut1 = [20, 62, +3.5]
cut2 = [15, 53, +3.5]
cut3 = [60, 42, +3.5]
cut4 = [111, 89,+3.5]

slice_3_3 = multicut(init_5_5, *zip(cut1,cut2,cut3,cut4))
#view(slice_3_3)

slice_3_3.write('slice_3_3.png')
Image(filename='slice_3_3.png')
```

```python
cut1 = [20, 62, +2.5]
cut2 = [15, 53, +2.5]
cut3 = [60, 42, +0.5]
cut4 = [111, 89,+0.5]

slice_4_4 = multicut(init_5_5, *zip(cut1,cut2,cut3,cut4))
#view(slice_4_4)

slice_4_4.write('slice_4_4.png')
Image(filename='slice_4_4.png')
```

```python
from ase.io import read
from ase.visualize import view
#created in avogadro aligned to the x-z plane
azo =  read('azo.xyz')
carb = read('carb.xyz')
```

```python
#between atoms 29 and 30
docking_position = 0.5*slice_3_3[12].position + 0.5*slice_3_3[13].position
#nano graphene orientated along x-y plane so offset 2.5 Angstroms in the z direction
docking_offset = np.array([0,0,2.5])
```

```python
carb.translate(docking_position + docking_offset)
for carb_at in carb:
    slice_3_3.append(carb_at)

slice_3_3.write('slice_3_3_carb.png')
Image(filename='slice_3_3_carb.png')
```
